```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv(r"C:\Users\Lenovo\Downloads\mobile_price_data.csv")
df
```

# data preprocessing

```python
#missing value check

df.isnull().sum()   # no missing values
```

```python
# feature encoding

#since all the cloumns are nominal in nature use one hot encoder 

df_ohe=pd.get_dummies(df[["mobile_name","mobile_color","dual_sim","disp_size","resolution","os","num_cores","mp_speed","int_memory","ram","p_cam","f_cam","network","bluetooth","battery_power"]])
df_ohe
```

```python
df=pd.concat([df,df_ohe],axis=1)
df
```

```python
df=df.drop(["mobile_name","mobile_color","dual_sim","disp_size","resolution","os","num_cores","mp_speed","int_memory","ram","p_cam","f_cam","network","bluetooth","battery_power"],axis=1)
```

```python
df=df.drop(["mob_width","mob_height","mob_depth","mob_weight"],axis=1)   # feature which are not  affect the price generally
df
```

# DIVIDE AND SPLIT DATA

```python
x=df.drop(["mobile_price"],axis=1)
y=df["mobile_price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# Apply logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
